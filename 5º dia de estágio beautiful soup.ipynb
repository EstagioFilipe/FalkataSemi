```python
import requests
from bs4 import BeautifulSoup

url = "https://www.minipreco.pt/produtos/frutas-e-vegetais/c/WEB.001.000.00000"
response = requests.get(url)

if response.status_code == 200:

    soup = BeautifulSoup(response.content, "html.parser")

    titulo = soup.find("h1").text
    paragrafos = soup.find_all("p")

    print("Título:", titulo)
    print("Parágrafos:")
    for p in paragrafos:
        print("-", p.text)
else:
    print("Falha ao acessar a página:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

def get_pagina(url):
    response = requests.get(url)
    return response.content

def extrair_dados_pagina(html):
    soup = BeautifulSoup(html, 'html.parser')
    links = soup.find_all('a')
    titulos = [link.text for link in links]
    return titulos

url = 'https://www.minipreco.pt/produtos/frutas-e-vegetais/c/WEB.001.000.00000'

html = get_pagina(url)

dados = extrair_dados_pagina(html)

for dado in dados:
    print(dado)
```

```python
import pandas as pd

req = urlopen(r'https://www.minipreco.pt/produtos/frutas-e-vegetais/c/WEB.001.000.00000').read()

soup = BeautifulSoup(req, "lxml")

prices = []
for links in soup.findAll('p', 'pricePerKilogram'):
    prices.append(str(links)[29:-6])

details = []
for links in soup.findAll('span', 'details'):
    details.append(str(links)[27:-7])

d = {'Produto': details[:-1], 'Price': prices}
lista = pd.DataFrame(data=d, index=None)

lista['Produto'] = lista['Produto'].str.replace('\t', '').str.replace('\n', '')
lista['Price'] = lista['Price'].str.replace(' €/Kg', ' ')
lista['Price'] = lista['Price'].str.replace(',', ' ')

lista['Unite'] = "€/Kg"
```
